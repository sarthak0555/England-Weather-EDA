```python
import pandas as pd
import itertools
```

```python
# Load the travel_distances file
travel_distances = pd.read_excel('/Users/sarthakchawla/Downloads/travel_distances.xlsx')
```

```python
travel_distances
```

```python
# Resetting the index to include 'From' column
travel_distances_reset = travel_distances.reset_index()
```

```python
# Restructuring the travel distances into a matrix format using the correct column names
travel_distances_matrix = travel_distances_reset.pivot(index='From', columns='To', values='TravelDistance')
```

```python
travel_distances_matrix
```

```python
# Define the cities in the problem
cities = ["London", "Birmingham", "Liverpool", "Leeds", "Sheffield", "Edinburgh", "Bristol", "Glasgow"]
```

```python
# Generate all permutations of the cities to be visited
to_visit = ["Birmingham", "Liverpool", "Leeds", "Sheffield", "Edinburgh", "Bristol"]
permutations = itertools.permutations(to_visit)
```

```python

# Function to calculate the total travel distance for a given route
def calculate_distance(route):
    total_distance = 0
    # Start in London
    current_city = "London"
    for city in route:
        total_distance += travel_distances_matrix.at[current_city, city]
        current_city = city
    # End in Glasgow
    total_distance += travel_distances_matrix.at[current_city, "Glasgow"]
    return total_distance
```

```python
# Find the route with the minimum total distance
min_distance = float('inf')
best_route = None

for route in permutations:
    distance = calculate_distance(route)
    if distance < min_distance:
        min_distance = distance
        best_route = route
```

```python
# Convert best route to a list starting with London and ending with Glasgow
best_route_list = ["London"] + list(best_route) + ["Glasgow"]
```
