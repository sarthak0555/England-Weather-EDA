```python
import numpy as np 
import pandas as pd
import matplotlib as mlt 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.cm import get_cmap
from matplotlib.patches import Patch
from matplotlib import rcParams
```

```python
df = pd.read_csv("/Users/sarthakchawla/Downloads/HRDataset_v14.csv.xls")
```

```python
df1 = df.groupby(['ManagerName'])['Termd'].value_counts().reset_index(name='count')
#df2=df2.sort_values('count')
df1 = df1.groupby(['ManagerName', 'Termd'])['count'].aggregate('first').unstack()
df1.to_csv('ques4.csv')
df1 = pd.read_csv('ques4.csv')
df1
```

```python
df1= df1.loc[:, df1.columns!='0']
df1.replace(np.nan, 0 , inplace=True)
df1 = df1.rename(columns={'1': 'No. of terminated employees'})
df1
```

```python
#df1.plot(kind='bar')
df1.plot(x='ManagerName', y='No. of terminated employees', kind='bar')
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.figure(figsize=(30,30))

plt.show()
```
