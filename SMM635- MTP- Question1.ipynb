```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HRDataset_v14.csv")
```

```python
df.head(20)
```

```python
df.info()
```

```python
managerdf = df[["ManagerName","PerfScoreID","PerformanceScore"]]
```

```python
newdf1 = managerdf.groupby(['ManagerName'])["PerformanceScore"].value_counts().reset_index(name='count')
newdf1
```

```python
x =newdf1.groupby(['ManagerName', 'PerformanceScore'])['count'].aggregate('first').unstack()
x.to_csv("new.csv")
```

```python
new = pd.read_csv("new.csv")
new.head(20)
```

```python
newdf = managerdf.groupby(['ManagerName'])["PerfScoreID"].mean().reset_index(name='mean')
newdf
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = new['ManagerName']

x = np.arange(len(labels)) 
y =  [i for i in range(20)]
y1 = np.arange(2.5,3.6,0.1)
fig = plt.figure(figsize=(15,8))
ax=fig.add_subplot(111)
line1 = ax.scatter(labels,new['Exceeds'], label='Exceeds',alpha=0.5,s=400)
line2 = ax.scatter(labels,new['Fully Meets'], label='Fully Meets', alpha = 0.5, s=300)
line3 = ax.scatter(labels,new['Needs Improvement'], label='Needs Improvement',alpha=0.6, s=170)
line4 = ax.scatter(labels,new['PIP'], label='PIP',alpha = 0.6, s=80)
ax.set_ylabel('Count')
ax.set_title('Performance score within and across Managers')
ax.set_xticks(x, labels)
ax.set_yticks(y)
ax.set_frame_on(False)
plt.xticks(rotation = 45)
fig.subplots_adjust(top = 2,bottom=1.5)
ax.legend( labelspacing =2,ncol=2, loc = 'upper left')

ax2 = ax.twinx()
ax2.plot(newdf['ManagerName'],newdf['mean'], marker = '*', color = 'black', alpha = 0.4, label = 'Average Rating')
ax2.set_ylabel('Average Rating',rotation=270,labelpad=15)
ax2.legend(loc= 'lower right')
ax2.set_frame_on(False)
ax2.set_yticks(y1)
for x,y in zip(newdf['ManagerName'],newdf['mean']):
    etiqueta = "{:.2f}".format(y)
    ax2.annotate(etiqueta,(x,y), textcoords="offset points",xytext=(0,10),ha="center")

fig.tight_layout()
plt.savefig('First.png')

plt.show()
```
