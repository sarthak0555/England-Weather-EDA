```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
ua = nx.read_graphml('ua.graphml')
```

```python
b = list(ua.edges(data=True))
```

```python
b
```

```python
e1 = []
u1 = []
for tup in b:
    e1.append(tup[0])
    u1.append(tup[1])
df1 = pd.DataFrame({'Employee':e1,
                  'Unit':u1
                   })    
df1
```

```python
pa = nx.read_graphml('pa.graphml')
```

```python
a = list(pa.edges(data=True))
a
```

```python
e2 = []
p1 = []
for tup in a:
    e2.append(tup[0])
    p1.append(tup[1])
df2 = pd.DataFrame({'project':p1,
                  'Employee':e2
                   })    
df2
```

```python
ie = nx.read_graphml('ie.graphml')
```

```python
n4 = list(ie.nodes(data = True))
n4
```

```python
e4 = list(ie.edges(data= True))
e4
```

```python
e3 = []
gender = []
ti_exp = []
tenure = []
for tup in n4:
    e3.append(tup[0])
    dic = tup[1]
    gender.append(dic['gender'])
    ti_exp.append(dic['ti_exp'])
    tenure.append(dic['tenure'])
df4 = pd.DataFrame({
                  'Employee':e3,
    'Gender' : gender,
    'ti_exp' : ti_exp,
    'Tenure' : tenure
                   })    
df4
```

```python

edge = []
strength = []
for tup in e4:
    edge.append(tup[0:2])
    dic = tup[2]
    strength.append(dic['strength'])
df5 = pd.DataFrame({
                  'Edge':edge,
    'Strength' : strength,
                   })    
df5
```

```python
edge_centrality = nx.edge_betweenness(ie)
```

```python
edge_centrality
```

```python
edge1 = []
ec = []
for k,v in edge_centrality.items():
    edge1.append(k)
    ec.append(v)
df6 = pd.DataFrame({
    'Edge': edge1,
    'Edge-centrality':ec
    
})    
df6
```

```python
edge_df = pd.merge(df5,df6,on='Edge')
edge_df
```
